```python
from faker import Faker
import random
import csv

fake = Faker()

# Real IEM brands and some of their popular models
iem_brands_models = {
    'Moondrop': ['Aria', 'Starfield', 'Blessing 2', 'KXXS', 'SSR', 'SSP', 'Chu', 'Kato'],
    'KZ': ['ZSN Pro X', 'ZS10 Pro', 'ZSX', 'ZAX', 'DQ6', 'EDX', 'CRN'],
    '7Hz': ['Timeless', 'Salnotes Zero', 'Salnotes Dioko'],
    'Tin HiFi': ['T2', 'T3', 'T4', 'P1', 'T2 Plus'],
    'FiiO': ['FH3', 'FD5', 'FA7', 'FH5', 'FD3'],
    'Shure': ['SE215', 'SE425', 'SE535', 'SE846'],
    'Etymotic': ['ER2XR', 'ER2SE', 'ER3XR', 'ER4XR'],
    'Campfire Audio': ['Andromeda', 'Solaris', 'IO', 'Honeydew', 'Mammoth'],
    'ThieAudio': ['Legacy 3', 'Legacy 4', 'Monarch', 'Clairvoyance'],
    'Fearless Audio': ['S8F', 'S8P', 'Roland'],
    'DUNU': ['SA6', 'EST112', 'DK-2001'],
    'Tanchjim': ['Oxygen', 'Hana', 'Tanya'],
    'Tripowin': ['TC-01', 'Mele', 'Olina'],
    'CCA': ['C10', 'C12', 'CRA', 'CRA+'],
    'BLON': ['BL-03', 'BL-05', 'BL-Mini'],
    'TRN': ['V90', 'BA5', 'ST1'],
    'Shozy': ['Form 1.1', 'Form 1.4'],
    'Ikko': ['OH1', 'OH10'],
    'Final Audio': ['E3000', 'E4000', 'E5000'],
    'Sony': ['IER-Z1R', 'IER-M9', 'MH755'],
}

sound_signatures = ['Flat', 'Balanced', 'Bright', 'V-Shaped', 'Extra Bass', 'Warm/Smooth']

def generate_model_name():
    return f"{fake.random_letter().upper()}{fake.random_int(min=1, max=99)}{fake.random_letter().upper()}"

data = []
for _ in range(10000):
    if random.random() < 0.7:  # 70% chance of using a real brand and model
        brand = random.choice(list(iem_brands_models.keys()))
        model = random.choice(iem_brands_models[brand])
    else:
        brand = fake.company()
        model = generate_model_name()

    year = random.randint(2010, 2024)
    impedance = random.randint(8, 64)
    soundstage = round(random.uniform(2.5, 5), 1)  # Assuming we're focusing on better IEMs
    detail_retrieval = round(random.uniform(2.5, 5), 1)
    sound_signature = random.choice(sound_signatures)
    build_quality = round(random.uniform(2.5, 5), 1)
    overall_rating = round(random.uniform(2.5, 5), 1)
    community_rating = round(random.uniform(2.5, 5), 1)

    data.append([brand, model, year, impedance, soundstage, detail_retrieval,
                 sound_signature, build_quality, overall_rating, community_rating])

# Write to CSV
with open('earphones_dataset.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Brand', 'Model', 'Year', 'Impedance', 'Soundstage', 'Detail Retrieval',
                     'Sound Signature', 'Build Quality', 'Overall Rating', 'Community Rating'])
    writer.writerows(data)

print("Dataset generated and saved as 'earphones_dataset.csv'")
```
